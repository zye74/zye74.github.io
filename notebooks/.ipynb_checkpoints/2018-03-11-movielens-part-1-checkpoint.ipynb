```python
# <!-- collapse=True -->
%matplotlib inline
from datetime import datetime
from IPython.display import Markdown
from os.path import exists
from os.path import join
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('big-darkgrid')
```

```python
# <!-- collapse=True -->
%%time
ratings_feather = '../preprocessed/ratings.feather'
ratings_csv = '../raw/ratings.csv'
if exists(ratings_feather):
    # Reading in feather file saves time
    ratings_df = pd.read_feather(ratings_feather)
else:
    ratings_df = pd.read_csv('../raw/ratings.csv')
    ratings_df = ratings_df.sort_values('timestamp').reset_index(drop=True)
    ratings_df['timestamp'] = ratings_df['timestamp'].apply(datetime.fromtimestamp)
    ratings_df['year'] = ratings_df['timestamp'].dt.year
    ratings_df['month'] = ratings_df['timestamp'].dt.month
    ratings_df['day'] = ratings_df['timestamp'].dt.day
    ratings_df['hour'] = ratings_df['timestamp'].dt.hour
    ratings_df['minute'] = ratings_df['timestamp'].dt.minute
    # Save to feather file for faster loading later
    ratings_df.to_feather(ratings_feather)
display(ratings_df.head())
```

# What is the distribution of the ratings?

```python
# <!-- collapse=True -->
ax = sns.countplot(x='rating', data=ratings_df)
ax.set_title('Rating Frequencies')
plt.show()
```

That's odd...why are the half-star scores less popular? Do people prefer whole numbers? Let's look at how this distribution changes over time to get more insight about that.

## Set up a rating frequency vs year pivot table to facilitate visualization

```python
ratings_pivot = pd.pivot_table(ratings_df, values='timestamp', index='rating', columns='year', aggfunc='count')
ratings_pivot = ratings_pivot.apply(lambda x: x/np.sum(x), axis=0)
ratings_pivot = ratings_pivot.sort_index(ascending=False)
ratings_pivot
```

```python
# <!-- collapse=True -->
fig, ax = plt.subplots(figsize=(15, 5))
sns.heatmap(ratings_pivot, ax=ax)
ax.set_title('Rating Frequency by Year\n(Sum over each year is 1)')
plt.show()
```

Ahh, this explains the rating distribution. It looks like half-stars weren't allowed until part-way through 2003. After 2003 the distribution looks pretty smooth and consistent. How do the average distributions compare before and after that change was made?

# What are the ratings distributions before and after half scores are allowed?

```python
# <!-- collapse=True -->
switch_timestamp = ratings_df[ratings_df['rating'].isin([0.5, 1.5, 2.5, 3.5, 4.5])].iloc[0]['timestamp']
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(12, 4))
sns.countplot(x='rating', data=ratings_df[ratings_df['timestamp']<switch_timestamp], ax=ax0)
sns.countplot(x='rating', data=ratings_df[ratings_df['timestamp']>=switch_timestamp], ax=ax1)
ax0.set_title('Rating Frequencies Pre Switch')
ax1.set_title('Rating Frequencies Post Switch')
plt.tight_layout()
plt.show()
```

# How consistent were ratings over time?

## First let's look at the number of ratings over time

```python
# <!-- collapse=True -->
fig, ax = plt.subplots(figsize=(15, 5))
sns.countplot(x='year', data=ratings_df, ax=ax)
ax.set_title('Number of Ratings Per Year')
plt.show()
n_1995 = np.sum(ratings_df['year']==1995)
print('{} ratings in 1995'.format(n_1995))
display(ratings_df[ratings_df['year']==1995])
```

The number of ratings has not been constant over the years. Setting aside the measley 4 (simultaneous down to the minute) ratings from 1995, (maybe test ratings at the very beginning of the project to make sure the system works?) there were large variations from 1996 to 2005 (possibly due to funding/grad studnet availability etc.?), then a steady decrease over time after that (maybe due to the rise of Netflix as a good movie recommender?).

## Now let's see how consistent the average rating was over the years

```python
%%time
ax = sns.regplot(data=ratings_df.sample(n=100000), x='year', y='rating', scatter=False)
# ax.set_ylim([0, 5])
ax.set_xlim([1995, 2015])
ax.set_xticks([1995, 2000, 2005, 2010, 2015])
ax.set_title('Rating vs Year Regression Plot')
plt.show()
```

This average ratings were extremely consistent around 3.5.

# How sparse is the user/movies matrix we'll be dealing with?

```python
user_counts = ratings_df['userId'].value_counts()
movie_counts = ratings_df['movieId'].value_counts()
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(14, 5))
ax0.plot(user_counts.values)
ax0.set_xlabel('User')
ax0.set_ylabel('# Movies Rated')
ax1.plot(movie_counts.values)
ax1.set_xlabel('Movie')
ax1.set_ylabel('# Users Rating')
plt.tight_layout()
plt.show()
```

The distribution of number of movies rated per user and users who rated each movie are both very long-tailed distributions, meaning most users rated very few movies, and most movies were rated very infrequently.

```python
num_users = ratings_df['userId'].nunique()
num_movies = ratings_df['movieId'].nunique()
num_ratings_given = len(ratings_df)
user_movies_matrix_density = num_ratings_given / num_users / num_movies
Markdown("""
We have {} users and {} movies, giving a possible total of {} ratings,
but we only have {} ratings, giving a matrix density of only {:7.5f}!
This is a very sparse matrix.
""".format(num_users, num_movies, num_users * num_movies, num_ratings_given, user_movies_matrix_density))
```
